```python
import tensorflow as tf
import numpy as np

# 1. Veri Гјretimi
# Normal: 36.4вЂ“36.6
normal = np.random.uniform(36.4, 36.6, 500)

# Anormal dГјЕџГјk: 34вЂ“35.5
anormal_low = np.random.uniform(34.0, 35.5, 250)

# Anormal yГјksek: 37.5вЂ“39.0
anormal_high = np.random.uniform(37.5, 39.0, 250)

# X giriЕџleri
X = np.concatenate([normal, anormal_low, anormal_high]).reshape(-1,1)

# y etiketleri: normal=1, anormal=0
y = np.concatenate([np.ones(500), np.zeros(500)])

# Normalize et (Г¶lГ§ekleme)
X = (X - 36.0) * 10

# 2. Model tanД±mД±
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 3. Model eДџitimi
model.fit(X, y, epochs=30, batch_size=32)

# 4. Test tahminleri (NumPy array ile!)
print("36.5 iГ§in tahmin:", model.predict(np.array([[ (36.5 - 36.0)*10 ]], dtype=np.float32)))
print("35.0 iГ§in tahmin:", model.predict(np.array([[ (35.0 - 36.0)*10 ]], dtype=np.float32)))
print("38.0 iГ§in tahmin:", model.predict(np.array([[ (38.0 - 36.0)*10 ]], dtype=np.float32)))

# 5. TFLite export
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("temp_model.tflite", "wb") as f:
    f.write(tflite_model)
```

```python
import numpy as np

# Test iГ§in farklД± sД±caklД±k deДџerleri
test_values = [35.0, 36.0, 36.5, 37.0, 38.0, 39.0,34.0,36.8,35.9,37.1]

for val in test_values:
    norm_val = (val - 36.0) * 10   # normalize et
    pred = model.predict(np.array([[norm_val]], dtype=np.float32))
    print(f"{val} В°C iГ§in tahmin: {pred[0][0]:.4f}")
```
